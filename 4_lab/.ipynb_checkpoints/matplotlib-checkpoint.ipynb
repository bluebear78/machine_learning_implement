```python
#data 시각화
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = range(100)
Y = [value**2 for value in X]
plt.plot(X,Y)
plt.show()
```

```python
import numpy as np
X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1,Y_1)
plt.plot(X_2,Y_2)
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10,5)
ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(X_1,Y_1,c='b')
ax_2.plot(X_2,Y_2,c='g')
plt.show()
```

```python
plt.plot(X_1,Y_1,color='b',linestyle='dashed',label = 'line_1')
plt.plot(X_2,Y_2,color='r',linestyle='dotted',label='line_2')

plt.grid(True,lw=0.4,ls='--',c=".90")
plt.xlim(-1000,2000) #scale
plt.ylim(-1000,2000)#yscale

plt.show()
```

```python
data_1 = np.random.rand(512,2)
data_2 = np.random.rand(512,2)

plt.scatter(data_1[:,0],data_1[:,1],c='b',marker='x')
plt.scatter(data_2[:,0],data_1[:,1],c='r',marker='^')

plt.show()
```

```python
data = [[5.,25.,50.,20.],
     [4.,23.,51.,17.],
     [6.,22.,52.,19]]
```

```python
X = np.arange(4)
plt.bar(X+0.00,data[0],color='b',width=0.25)
plt.bar(X+0.25,data[1],color='g',width=0.25)
plt.bar(X+0.50,data[2],color='r',width=0.25)
plt.xticks(X+0.25,("A","B","C","D"))
plt.show()
```

```python
data = np.array(data)
color_list = ['b','g','r']
data_label = ["A","B","C"]

X = np.arange(data.shape[1])
for i in range(data.shape[0]):
    plt.bar(X,data[i],bottom = np.sum(data[:i],axis=0),
           color = color_list[i],label=data_label[i])
plt.legend()
plt.show()
```

```python
women_pop = np.array([5,30,45,22])
men_pop = np.array([5,25,50,20])

X = np.arange(4)

plt.barh(X,women_pop,color='r')
plt.barh(X,-men_pop,color='b')
plt.show()
```
