```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data_url = './BostonHousing.csv'
df_data = pd.read_csv(data_url,sep=',',header = None)
df_data.head()
```

```python
df_data = df_data.rename(columns=df_data.iloc[0])
df_data.drop(df_data.index[0],inplace=True)
df_data.head()
```

```python
fig = plt.figure()
ax = []
for i in range(1,5):
    ax.append(fig.add_subplot(2,2,i))
ax[0].scatter(df_data['CRIM'],df_data['MEDV'])
ax[1].scatter(df_data['PTRATIO'],df_data['MEDV'])
ax[2].scatter(df_data['DIS'],df_data['MEDV'])
ax[3].scatter(df_data['NOX'],df_data['MEDV'])
plt.show()
```

```python
fig = plt.figure()
ax = []
for i in range(1,5):
    ax.append(fig.add_subplot(2,2,i))
columns = ["CRIM","PTRATIO","AGE ","NOX"]
colors = ['b','g','c','r']


for i in range(4):
    ax[i].scatter(df_data[columns[i]],df_data["MEDV"],color=colors[i],label = columns[i])
    ax[i].axes.xaxis.set_visible(False)
    ax[i].axes.yaxis.set_visible(False)
    ax[0].legend()
    ax[0].set_title("CRIM")
plt.subplots_adjust(wspace=0,hspace=0)
plt.show()
```

```python
corr_data = np.corrcoef(df_data)
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr_data,vmin=-1,vmax=1,interploation='nearest')
fig.colorbar(cax)
fig.set_size_inches(10,10)
ticks = np.arange(0,14,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(df_data.columns)
ax.set_yticklabels(df_data.columns)
```
