```python
import dateutil
import pandas as pd
```

```python
df_phone = pd.read_csv("phone_data.csv")
df_phone.head()
```

```python
df_phone['date'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst=True)
df_phone.head()
```

```python
df_phone.groupby('month')['duration'].sum()
```

```python
df_phone.month.unique()
```

```python
df_phone[df_phone['item']=='call'].groupby('network')['duration'].sum()
```

```python
df_phone[df_phone['item']=='call'].groupby('month')['duration'].sum()
```

```python
df_phone.groupby(['item','month'])['duration'].sum()
```

```python
df_phone.groupby(['month','item'])['duration'].sum()
```

```python
df_phone.groupby(['month','item'])['date'].count().unstack()
```

```python
df_phone.groupby('month',as_index=False).agg({'duration':"mean"})
```

```python
df_phone.groupby(['month','item'],as_index=False).agg({'duration':sum, 'network_type':'count','date':'first'})
```

```python
df_phone.groupby(['month','item']).agg({'duration':[min,max,sum], 'network_type':'count','date':[min,'first','nunique']})
```

```python
import numpy as np
grouped = df_phone.groupby('month').agg({"duration" : [min,max,np.mean]})
grouped.head()
```

```python
grouped.columns = grouped.columns.droplevel(level=0)
```

```python
grouped.head()
```

```python
grouped.rename(columns={'min':'min_duration','max':'max_duration','mean':'mean_duration'})
```

```python
grouped.add_prefix("duration_")
```

```python
df_phone.pivot_table(['duration'],index = [df_phone.month,df_phone.item], columns = df_phone.network,aggfunc="sum",fill_value = 0)
```
